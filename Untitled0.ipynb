```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dates = pd.date_range(start="2020-01-01", end="2021-12-31")

np.random.seed(0)
deaths = np.random.poisson(lam=10, size=len(dates)).cumsum()


covid_data = pd.DataFrame({"Date": dates, "Deaths": deaths})


print(covid_data.head())

plt.figure(figsize=(10, 6))
plt.plot(covid_data["Date"], covid_data["Deaths"], label="Mortes por COVID-19", color='red')
plt.fill_between(covid_data["Date"], covid_data["Deaths"], alpha=0.3, color='red')
plt.title("Nmero Cumulativo de Mortes por COVID-19")
plt.xlabel("Data")
plt.ylabel("Nmero de Mortes")
plt.legend()
plt.grid(True)
plt.show()
```
